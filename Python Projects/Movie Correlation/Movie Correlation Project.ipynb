# Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline

#Adjusting the configuration of the plots we will create
matplotlib.rcParams['figure.figsize'] = (12,8) 
```

# Read in the data

```python
df = pd.read_csv(r'D:\New Projects\Movie Data\movies.csv')
```

# Checking for Missisng Data

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
    
```

# Cleaning the Data

```python
#Changing data types of columns

df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.min_rows', None)
```

```python
#Droping any duplicate values

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.drop_duplicates()
```

# Building a Scatter Plot

```python
#Scatter plot with budget vs gross

df.plot(kind='scatter',x='budget',y='gross',color='purple')
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Movie')
plt.show()
```

```python
df.head()
```

```python
#Checking for outliers

df.boxplot(column=['gross'])
```

```python
#Plotting Budget vs gross using seaborn

df['budget'] = df['budget'].astype('float64')

df['gross'] = df['gross'].astype('float64')

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "orange"}, line_kws={"color": "purple"})
```

```python
#Plotting Score vs gross using seaborn
sns.regplot(x="score", y="gross", data=df, scatter_kws={"color": "orange"}, line_kws={"color": "purple"})
```

```python
#Plotting Budget vs score using seaborn
sns.regplot(x="budget", y="score", data=df, scatter_kws={"color": "orange"}, line_kws={"color": "purple"})
```

# Visualizing Correlation Matrix

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

# numerizing company column

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
    
```

```python
df.head()
```

# Visualizing Correlation Matrix for Numerized column 

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

# Filtering the Correlation matrix for better Visualiztion

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

Besides budget and gross, Votes and gross are also highly correlated
